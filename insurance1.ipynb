Generating data to play with them

Play of Herbivorous and carnivorous


```python
valoresP = {"name": ['koala', 'caterpillar', 'deer', 'human'],
            "id": [1,2,3,4],
            "eats_plants": ['yes', 'yes', 'yes', 'yes']}
eats_plants = pd.DataFrame(valoresP)

valoresM = {"name": ['shark', 'lion', 'tiger', 'human'],
            "id": [5,6,7,4],
            "eats_meats": ['yes', 'yes', 'yes', 'yes']}
eats_meat = pd.DataFrame(valoresM)


```

```python
print("Eat plants (herbivorous)")
print(eats_plants)
print("-------------------\n")

print("Eat meat (carnivorous)")
print(eats_meat)

```

```python
# Append

new_df = eats_plants.copy()
new_df = pd.concat([eats_plants, new_df])
new_df
```

```python
# Inner join

# inner

inner_join = eats_plants.merge(eats_meat, on= ['name', 'id'], how="inner")
inner_join
```

```python
eats_plants.merge(eats_meat, on= ['id'], how="inner")
```

```python
# Inner join

# left

left_join = eats_plants.merge(eats_meat, on= ['name', 'id'], how="left")
left_join
```

```python
# Inner join

# right

right_join = eats_plants.merge(eats_meat, on= ['name', 'id'], how="right")
right_join
```

```python
full_outer_join = eats_plants.merge(eats_meat, on=['name', 'id'], how='outer')
full_outer_join
```
